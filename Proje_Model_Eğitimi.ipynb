```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.regularizers import l2
import matplotlib.pyplot as plt

# ============================================================
# VERÄ° YOLU (Ã–nceki koddan devam)
# ============================================================
balanced_dir = '/content/balanced_dataset'

# Veri setini kontrol et
if not os.path.exists(balanced_dir):
    print("âŒ Dengeli veri seti bulunamadÄ±!")
    print("LÃ¼tfen Ã¶nce Ã¶nceki kodu Ã§alÄ±ÅŸtÄ±rÄ±n.")
    raise FileNotFoundError("balanced_dataset klasÃ¶rÃ¼ yok!")

print("=" * 60)
print("VERÄ° SETÄ° KONTROLDEKÄ° SINIFLAR")
print("=" * 60)
for class_name in os.listdir(balanced_dir):
    class_path = os.path.join(balanced_dir, class_name)
    if os.path.isdir(class_path):
        count = len([f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.png', '.jpeg'))])
        print(f"{class_name:25s}: {count:5d} gÃ¶rÃ¼ntÃ¼")

# ============================================================
# VERÄ° YÃœKLEME - RGB Ä°Ã§in (Transfer Learning)
# ============================================================
# MobileNetV2 RGB gÃ¶rÃ¼ntÃ¼ bekliyor (3 kanal)
img_height, img_width = 96, 96  # Daha yÃ¼ksek Ã§Ã¶zÃ¼nÃ¼rlÃ¼k
batch_size = 32

# EÄŸitim augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.15,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest',
    validation_split=0.25
)

# Validation sadece rescale
val_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.25
)

train_generator = train_datagen.flow_from_directory(
    balanced_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    color_mode="rgb",  # RGB'ye Ã§evir (gri deÄŸil)
    class_mode='categorical',
    subset='training',
    shuffle=True
)

validation_generator = val_datagen.flow_from_directory(
    balanced_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    color_mode="rgb",
    class_mode='categorical',
    subset='validation',
    shuffle=False
)

num_classes = len(train_generator.class_indices)
print(f"\nâœ“ EÄŸitim: {train_generator.samples} gÃ¶rÃ¼ntÃ¼")
print(f"âœ“ DoÄŸrulama: {validation_generator.samples} gÃ¶rÃ¼ntÃ¼")
print(f"âœ“ SÄ±nÄ±f sayÄ±sÄ±: {num_classes}")
print(f"âœ“ SÄ±nÄ±flar: {list(train_generator.class_indices.keys())}")

# ============================================================
# TRANSFER LEARNING MODELÄ° (MobileNetV2)
# ============================================================
print("\n" + "=" * 60)
print("TRANSFER LEARNING MODELÄ° OLUÅTURULUYOR...")
print("=" * 60)

# Pre-trained MobileNetV2 (ImageNet aÄŸÄ±rlÄ±klarÄ± ile)
base_model = MobileNetV2(
    input_shape=(img_height, img_width, 3),
    include_top=False,  # Son katmanlarÄ± ekleme
    weights='imagenet'   # ImageNet aÄŸÄ±rlÄ±klarÄ±
)

# Ä°lk katmanlarÄ± dondur (fine-tuning iÃ§in)
base_model.trainable = False

print(f"âœ“ Base model yÃ¼klendi: {base_model.name}")
print(f"âœ“ Base model katman sayÄ±sÄ±: {len(base_model.layers)}")

# Kendi sÄ±nÄ±flandÄ±rma katmanlarÄ±nÄ± ekle
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    BatchNormalization(),
    Dense(256, activation='relu', kernel_regularizer=l2(0.001)),
    Dropout(0.5),
    BatchNormalization(),
    Dense(128, activation='relu', kernel_regularizer=l2(0.001)),
    Dropout(0.4),
    Dense(num_classes, activation='softmax')
])

model.compile(
    optimizer=Adam(learning_rate=0.001),  # Ä°lk aÅŸama iÃ§in daha hÄ±zlÄ±
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\n" + "=" * 60)
print("MODEL YAPISI")
print("=" * 60)
model.summary()

# ============================================================
# AÅAMA 1: DONDURULMUÅ BASE MODEL Ä°LE EÄÄ°TÄ°M
# ============================================================
print("\n" + "=" * 60)
print("AÅAMA 1: Base Model DondurulmuÅŸ EÄŸitim (HÄ±zlÄ±)")
print("=" * 60)

early_stop_1 = EarlyStopping(
    monitor='val_accuracy',
    patience=10,
    restore_best_weights=True,
    mode='max',
    verbose=1
)

reduce_lr_1 = ReduceLROnPlateau(
    monitor='val_accuracy',
    factor=0.5,
    patience=3,
    min_lr=1e-7,
    mode='max',
    verbose=1
)

checkpoint_1 = ModelCheckpoint(
    'transfer_phase1.h5',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

history_1 = model.fit(
    train_generator,
    epochs=30,
    validation_data=validation_generator,
    callbacks=[early_stop_1, reduce_lr_1, checkpoint_1],
    verbose=1
)

# ============================================================
# AÅAMA 2: FINE-TUNING (Son KatmanlarÄ± AÃ§)
# ============================================================
print("\n" + "=" * 60)
print("AÅAMA 2: Fine-Tuning (Son Katmanlar AÃ§Ä±lÄ±yor)")
print("=" * 60)

# Base modelin son 30 katmanÄ±nÄ± eÄŸitilebilir yap
base_model.trainable = True
fine_tune_at = len(base_model.layers) - 30

for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

# Daha yavaÅŸ Ã¶ÄŸrenme hÄ±zÄ± ile yeniden derle
model.compile(
    optimizer=Adam(learning_rate=0.0001),  # Daha yavaÅŸ
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print(f"âœ“ EÄŸitilebilir katmanlar: {sum([1 for layer in model.layers if layer.trainable])}")

early_stop_2 = EarlyStopping(
    monitor='val_accuracy',
    patience=15,
    restore_best_weights=True,
    mode='max',
    verbose=1
)

reduce_lr_2 = ReduceLROnPlateau(
    monitor='val_accuracy',
    factor=0.5,
    patience=5,
    min_lr=1e-8,
    mode='max',
    verbose=1
)

checkpoint_2 = ModelCheckpoint(
    'transfer_best_model.h5',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

history_2 = model.fit(
    train_generator,
    epochs=50,
    validation_data=validation_generator,
    callbacks=[early_stop_2, reduce_lr_2, checkpoint_2],
    verbose=1
)

# ============================================================
# SONUÃ‡LAR VE GRAFÄ°KLER
# ============================================================
model.save('final_transfer_model.h5')

# TÃ¼m geÃ§miÅŸi birleÅŸtir
all_history = {
    'accuracy': history_1.history['accuracy'] + history_2.history['accuracy'],
    'val_accuracy': history_1.history['val_accuracy'] + history_2.history['val_accuracy'],
    'loss': history_1.history['loss'] + history_2.history['loss'],
    'val_loss': history_1.history['val_loss'] + history_2.history['val_loss']
}

# En iyi sonuÃ§lar
best_idx = np.argmax(all_history['val_accuracy'])

print("\n" + "=" * 60)
print("TRANSFER LEARNING SONUÃ‡LARI")
print("=" * 60)
print(f"\nEn Ä°yi Epoch: {best_idx + 1}")
print(f"EÄŸitim DoÄŸruluÄŸu: {all_history['accuracy'][best_idx]:.4f} ({all_history['accuracy'][best_idx]*100:.2f}%)")
print(f"DoÄŸrulama DoÄŸruluÄŸu: {all_history['val_accuracy'][best_idx]:.4f} ({all_history['val_accuracy'][best_idx]*100:.2f}%)")
print(f"EÄŸitim KaybÄ±: {all_history['loss'][best_idx]:.4f}")
print(f"DoÄŸrulama KaybÄ±: {all_history['val_loss'][best_idx]:.4f}")

gap = all_history['accuracy'][best_idx] - all_history['val_accuracy'][best_idx]
print(f"\nGenelleme FarkÄ±: {gap:.4f} ({gap*100:.2f}%)")

if gap < 0.05:
    print("âœ… MÃ¼kemmel genelleme!")
elif gap < 0.10:
    print("âœ“ Ä°yi genelleme")
elif gap < 0.15:
    print("âš ï¸ Kabul edilebilir")
else:
    print("âŒ Overfitting var")

# Grafik Ã§iz
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# DoÄŸruluk grafiÄŸi
axes[0].plot(all_history['accuracy'], label='EÄŸitim')
axes[0].plot(all_history['val_accuracy'], label='DoÄŸrulama')
axes[0].axvline(x=len(history_1.history['accuracy']), color='r', linestyle='--', label='Fine-tuning baÅŸlangÄ±cÄ±')
axes[0].set_title('Model DoÄŸruluÄŸu')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('DoÄŸruluk')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# KayÄ±p grafiÄŸi
axes[1].plot(all_history['loss'], label='EÄŸitim')
axes[1].plot(all_history['val_loss'], label='DoÄŸrulama')
axes[1].axvline(x=len(history_1.history['loss']), color='r', linestyle='--', label='Fine-tuning baÅŸlangÄ±cÄ±')
axes[1].set_title('Model KaybÄ±')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('KayÄ±p')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('training_history.png', dpi=150, bbox_inches='tight')
print("\nğŸ“Š Grafik kaydedildi: training_history.png")
plt.show()

print("\nğŸ’¾ Modeller:")
print("   âœ“ transfer_best_model.h5 (EN Ä°YÄ°)")
print("   âœ“ final_transfer_model.h5 (SON)")
print("   âœ“ transfer_phase1.h5 (AÅŸama 1)")

# Performans karÅŸÄ±laÅŸtÄ±rmasÄ±
print("\n" + "=" * 60)
print("PERFORMANS KARÅILAÅTIRMASI")
print("=" * 60)
print(f"Ã–nceki CNN Model:    ~%40 doÄŸruluk")
print(f"Transfer Learning:   {all_history['val_accuracy'][best_idx]*100:.1f}% doÄŸruluk")
print(f"Ä°yileÅŸme:            +{(all_history['val_accuracy'][best_idx] - 0.40)*100:.1f}%")
```

### Model Performans DeÄŸerlendirmesi

Modelin daha detaylÄ± performansÄ±nÄ± gÃ¶rmek iÃ§in doÄŸruluk (accuracy), kesinlik (precision), geri Ã§aÄŸÄ±rma (recall) ve F1-skoru gibi metrikleri hesaplayalÄ±m. Bunun iÃ§in modelimizi yÃ¼kleyip doÄŸrulama seti Ã¼zerinde tahminler yapacaÄŸÄ±z.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, confusion_matrix

# ============================================================
# AYARLAR
# ============================================================
model_path = 'transfer_best_model.h5'  # En iyi modeli yÃ¼klÃ¼yoruz
data_dir = '/content/balanced_dataset' # Veri seti yolu
img_height, img_width = 96, 96
batch_size = 32

# ============================================================
# 1. VERÄ°YÄ° HAZIRLA (SHUFFLE=FALSE Ã‡OK Ã–NEMLÄ°)
# ============================================================
# Test/DoÄŸrulama verisini aynen yÃ¼klÃ¼yoruz ama KARIÅTIRMIYORUZ
val_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.25
)

print("Veriler yÃ¼kleniyor...")
val_generator = val_datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    color_mode='rgb',      # Transfer learning iÃ§in RGB ÅŸart
    subset='validation',
    shuffle=False          # DÄ°KKAT: Rapor iÃ§in False olmalÄ±!
)

class_names = list(val_generator.class_indices.keys())
print(f"SÄ±nÄ±flar: {class_names}")

# ============================================================
# 2. MODELÄ° YÃœKLE VE TAHMÄ°N ET
# ============================================================
if os.path.exists(model_path):
    print(f"\nModel yÃ¼kleniyor: {model_path}...")
    model = load_model(model_path)
    print("Model baÅŸarÄ±yla yÃ¼klendi.")
else:
    raise FileNotFoundError("Model dosyasÄ± bulunamadÄ±! LÃ¼tfen Ã¶nce eÄŸitimi tamamlayÄ±n.")

print("\nTahminler yapÄ±lÄ±yor (Bu iÅŸlem biraz sÃ¼rebilir)...")
# Modelden olasÄ±lÄ±klarÄ± al (Ã–rn: [%10, %80, %5, %2, %3])
predictions = model.predict(val_generator, verbose=1)

# En yÃ¼ksek olasÄ±lÄ±klÄ± sÄ±nÄ±fÄ± seÃ§ (Ã–rn: 1. index -> KÄ±zgÄ±n)
y_pred = np.argmax(predictions, axis=1)

# GerÃ§ek etiketleri al
y_true = val_generator.classes

# ============================================================
# 3. SINIFLANDIRMA RAPORU
# ============================================================
print("\n" + "="*60)
print("SINIFLANDIRMA RAPORU")
print("="*60)
print(classification_report(y_true, y_pred, target_names=class_names))

# ============================================================
# 4. KARMAÅIKLIK MATRÄ°SÄ° (CONFUSION MATRIX)
# ============================================================
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names,
            yticklabels=class_names)

plt.title('KarmaÅŸÄ±klÄ±k Matrisi (Confusion Matrix)')
plt.ylabel('GerÃ§ek SÄ±nÄ±flar')
plt.xlabel('Tahmin Edilen SÄ±nÄ±flar')
plt.show()

# ============================================================
# 5. HATALI TAHMÄ°NLERÄ° GÃ–STER (ANALÄ°Z Ä°Ã‡Ä°N)
# ============================================================
# Modelin en Ã§ok nerede hata yaptÄ±ÄŸÄ±nÄ± gÃ¶rmek iÃ§in Ã¶rnekler
errors = np.where(y_pred != y_true)[0]
print(f"\nToplam HatalÄ± Tahmin SayÄ±sÄ±: {len(errors)}")

if len(errors) > 0:
    print("\nHatalÄ± tahminlerden 5 Ã¶rnek:")
    plt.figure(figsize=(15, 5))

    # Rastgele 5 hatayÄ± seÃ§
    random_errors = np.random.choice(errors, min(5, len(errors)), replace=False)

    for i, error_idx in enumerate(random_errors):
        img_path = val_generator.filepaths[error_idx]
        img = plt.imread(img_path)

        true_label = class_names[y_true[error_idx]]
        pred_label = class_names[y_pred[error_idx]]
        confidence = np.max(predictions[error_idx])

        plt.subplot(1, 5, i+1)
        plt.imshow(img)
        plt.title(f"GerÃ§ek: {true_label}\nTahmin: {pred_label}\n(GÃ¼ven: {confidence:.2f})", color='red', fontsize=9)
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```
